***Описание датасета***

Датасет "house" состоит из 7 столбцов, к которым относятся:


1.   index - порядковый номер позиции в таблице;
2.   SquareFeet - квадратный фут (единица измерения площади в английской системе мер, «международный» фут стал равняться в точности 0,3048 м);
3.   Bedrooms - количество спальных комнат;
4.   Bathrooms - количество ванных комнат;
5.   Neighborhood - район (Rural - сельский, Suburb - пригородный, Urban - городской);
6.   YearBuilt - год постройки;
7.   Price - стоимость дома ($).

```python
#Преобразование столбца "Price" из вещественного в целочисленный формат
df['Price'] = df['Price'].astype (int)
df.info()
df
```

Предположим, что мы хотим найти среднюю цену домов, построенных после 2000 года, с квадратурой от 1200 до 1800 квадратных футов и с количеством спальных комнат не менее 3. Для этого мы можем использовать map для преобразования данных и reduce для вычисления средней цены.

```python
# Функция Map
def map_function(df):
    filtered_prices = []
    for index, row in df.iterrows():
        if (row["YearBuilt"] > 2000 and
            1200 <= row["SquareFeet"] <= 1800 and
            row["Bedrooms"] >= 3):
            filtered_prices.append(row["Price"])
    return filtered_prices

# Функция Reduce
def reduce_function(mapped_data):
    if not mapped_data:
        return None
    total_price = sum(mapped_data)
    count_houses = len(mapped_data)
    return total_price / count_houses if count_houses > 0 else None

# Применяем функции Map и Reduce
mapped_data = map_function(df)
average_price = reduce_function(mapped_data)

print("Средняя цена домов, построенных после 2000 года, с квадратурой от 1200 до 1800 квадратных футов и с количеством спален не менее 3:",
      round(average_price,2),"долларов.")
```
