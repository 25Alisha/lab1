***Описание датасета***

Датасет "house" состоит из 7 столбцов, к которым относятся:


1.   index - порядковый номер позиции в таблице;
2.   SquareFeet - квадратный фут (единица измерения площади в английской системе мер, «международный» фут стал равняться в точности 0,3048 м);
3.   Bedrooms - количество спальных комнат;
4.   Bathrooms - количество ванных комнат;
5.   Neighborhood - район (Rural - сельский, Suburb - пригородный, Urban - городской);
6.   YearBuilt - год постройки;
7.   Price - стоимость дома ($).

Импорт библиотек и чтение датасета

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
df = pd.read_csv('drive/MyDrive/house.csv')
df=df.reset_index()
df
```

```python
df.info()
```

```python
#Преобразование столбца "Price" из вещественного в целочисленный формат
df['Price'] = df['Price'].astype (int)
df.info()
df
```

```python
# Чтобы использовать функции map и reduce из модуля functools,импортируем их:
from functools import reduce
from operator import attrgetter
```

Предположим, что мы хотим найти среднюю цену домов, построенных после 2000 года, с квадратурой от 1200 до 1800 квадратных футов и с количеством спальных комнат не менее 3. Для этого мы можем использовать map для преобразования данных и reduce для вычисления средней цены.

```python
filtered_data = list(
    map(
        lambda house: house[1]['Price'],
        filter(
            lambda house: house[1]["YearBuilt"] > 2000 and 1200 <= house[1]["SquareFeet"] <= 1800 and house[1]["Bedrooms"] >= 3,
            df.iterrows()
        )
    )
)

total_price = reduce(lambda x, y: x + y, filtered_data)
average_price = total_price / len(filtered_data)

print("Цены домов, удовлетворяющих параметрам:", filtered_data)
print("Средняя цена домов, построенных после 2000 года, с квадратурой от 1200 до 1800 квадратных футов и с количеством спальных комнат не менее 3 равна", round(average_price, 2), "$")
```
